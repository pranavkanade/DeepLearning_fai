```
import tensorflow as tf
```

```
x1 = tf.constant(5)
x2 = tf.constant(6)
```

```
# we can get away with this
result_simp = x1*x2
print (result_simp)
```

```
# original way of doing the multiplication
result = tf.multiply(x1, x2)    # mul is no longer in tensorflow
print(result)   # is the abstract tensor
```

```
sess = tf.Session()
print(sess.run(result))
sess.close()
```

```
# try to do this always over the above method
with tf.Session() as sess:
  print(sess.run(result))
```
